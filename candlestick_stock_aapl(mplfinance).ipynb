```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as fplt
```

### **Data Cleaning**

```python
df = pd.read_csv('/content/all_stocks_5yr.csv')
df.head()
```

**Checking type of Data**

```python
df.dtypes
```

**Change date type from object type to date type.**

```python
df['date'] = pd.to_datetime(df['date'],format='%Y/%m/%d')
```

```python
df.dtypes
```

**Replace the column name from name to ticks.**

```python
df = df.rename(columns={'Name':'Tick'})
df
```

**Check null values**

```python
null = df.isnull().sum()
print(null)
```

```python
#drop null values
df.dropna(inplace=True)
```

```python
null = df.isnull().sum()
print(null)
```

### **Data Analysis**

Analysis aapl stock

*   analysis aapl 2017
*   additional EMA calculate


**Select data only aapl stock**

```python
aapl = df.loc[df['Tick'] == 'AAPL']
aapl.head()
```

**EMA Calculate**

```python
# SettingWithCopyError
pd.options.mode.copy_on_write = True

aapl['EMA_20'] = aapl.loc[:,'close'].ewm(span=20, adjust = False).mean()
aapl['EMA_50'] = aapl.loc[:,'close'].ewm(span=50, adjust = False).mean()
aapl['EMA_100'] = aapl.loc[:,'close'].ewm(span=100, adjust = False).mean()
aapl['EMA_200'] = aapl.loc[:,'close'].ewm(span=200, adjust = False).mean()

aapl
```

```python
mask = (aapl['date'] >= '2017-01-01') & (aapl['date'] <= '2017-12-31')
aapl_2017 = aapl.loc[mask]
aapl_2017
```

**Determine date be index for this dataframe**

```python
aapl_2017.set_index(aapl_2017['date'],inplace=True)
```

**Create graph AAPL**

```python
## Adding data to candlestick pattern subplot
add_plots = {
    "EMA_20" : fplt.make_addplot(aapl_2017['EMA_20']),
    "EMA_50" : fplt.make_addplot(aapl_2017['EMA_50']),
    "EMA_100": fplt.make_addplot(aapl_2017['EMA_100']),
    "EMA_200" : fplt.make_addplot(aapl_2017['EMA_200'])
}
```

```python
fig, axes = fplt.plot(aapl_2017,
                    type='candle',
                    style='charles',
                    addplot=list(add_plots.values()),
                    returnfig=True,
                    title = 'Exponential moving average for Apple',
                    ylabel = 'Price($)',
                    volume = True)
# Adding legend
axes[0].legend([None]*(len(add_plots)+2))
handles = axes[0].get_legend().legend_handles
axes[0].legend(handles=handles[2:], labels=list(add_plots.keys()), loc='upper left', fontsize='x-small')
```
