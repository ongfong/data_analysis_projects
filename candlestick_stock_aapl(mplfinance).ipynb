```python
import pandas as pd
import matplotlib.pyplot as plt
import mplfinance as fplt
```

```python
df = pd.read_csv('/content/all_stocks_5yr.csv')
df.head()
```

```python
df.dtypes
```

```python
df['date'] = pd.to_datetime(df['date'],format='%Y/%m/%d')
```

```python
df.dtypes
```

```python
df = df.rename(columns={'Name':'Tick'})
df
```

```python
null = df.isnull().sum()
print(null)
```

```python
df.dropna(inplace=True)
```

```python
null = df.isnull().sum()
print(null)
```

```python
aapl = df.loc[df['Tick'] == 'AAPL']
aapl.head()
```

```python
aapl['EMA_20'] = aapl['close'].ewm(span=20, adjust=False).mean()
aapl['EMA_50'] = aapl['close'].ewm(span=50, adjust=False).mean()
aapl['EMA_100'] = aapl['close'].ewm(span=100, adjust=False).mean()
aapl['EMA_200'] = aapl['close'].ewm(span=200, adjust=False).mean()
aapl.head()
```

```python
mask = (aapl['date'] >= '2017-01-01') & (aapl['date'] <= '2017-12-31')
aapl_2017 = aapl.loc[mask]
aapl_2017
```

```python
aapl_2017.set_index(aapl_2017['date'],inplace=True)
```

```python
## Adding data to candlestick pattern subplot
ema_20 = fplt.make_addplot(aapl_2017['EMA_20'],color='red')
ema_50 = fplt.make_addplot(aapl_2017['EMA_50'],color='green')
ema_100 = fplt.make_addplot(aapl_2017['EMA_100'],color='purple')
ema_200 = fplt.make_addplot(aapl_2017['EMA_200'],color='blue')

fplt.plot(
    aapl_2017,
    type='candle',
    style='charles',
    addplot=[ema_20,ema_50,ema_100,ema_200],
    figratio=(20,6),
    title = 'AAPL Price 2017(with EMA20,EMA50,EMA100,EMA200)',
    ylabel = 'Price($)',
    volume = True,
)
```
